<p><b> iii. For address column. </b></p>

<p>We have <b>'latitude'</b> and <b>'longitude'</b> column.</p>
<p>We will use <b> geopy</b> library to impute the missing values in this column.</p>

```python
!pip install geopy
```

```python
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
# Creating a geocoder object with a higher timeout
geolocator = Nominatim(user_agent="restaurant_geocoder", timeout=10)
# Function to get address from latitude and longitude
def get_address(lat, lon):
    location = geolocator.reverse((lat, lon))
    return location.address if location else None

# Assuming df is your DataFrame and 'latitude', 'longitude', and 'address' are column names
# Fill missing addresses based on latitude and longitude
df_2['address'] = df_2.apply(lambda row: row['address'] if pd.notnull(row['address']) else get_address(row['latitude'], row['longitude']), axis=1)
```

<p><b> iv. For timings column. </b></p>

<p>And, as <b>timings</b> is a categorical column.</p>

<p> Ways to <b>impute/deal</b> with missing values in a categorical column.</p>
    <p> i.  <b>Delete the missing values record: </b> Deleting a whole record where timing has a missing value is not sufficient.</p>
    <p> ii. <b>Filling with most frequent value: </b> Filling missing values with a mode in categorical column can lead to biasness. Hence, we will not use mode here.</p>
    <p> iii.  <b>Imputing with placeholder: </b> Instead of deleting or imputing with mode, we will use a placeholder to impute it.</p>
    
   <p>When we have a <b>categorical column</b>, it is best to fill the missing values by applying a <b>classifier   
   algorithm</b> to predict missing values.</p>
   <p> As we have not studied studied yet. We will use placeholder instead </p>    


```python
df_2['timings'] = df_2['timings'].fillna('Not Available')
```
